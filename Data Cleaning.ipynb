```python
import pandas as pd
```

```python
#Import CSV files
csv_file = "Resources/athletes.csv"
athletes_df = pd.read_csv(csv_file)
athletes_df.head()
```

```python
#Set the data to only show USA gold athletes
usa_athletes = athletes_df.loc[athletes_df["nationality"]== 'USA']
gold_usa = usa_athletes.loc[usa_athletes["gold"]== 1]
gold_usa
```

```python
#Import CSV files
csv_file = "Resources/Summer-Olympic-medals-1976-to-2008.csv"
summer_df = pd.read_csv(csv_file)
summer_df.head()
```

```python
#Set the data to only show USA gold athletes
usa_summer_athletes = summer_df.loc[summer_df["Country_Code"]== 'USA']
summer_gold_usa = usa_summer_athletes.loc[usa_summer_athletes["Medal"]== 'Gold']
summer_gold_usa
```

```python
#Remove silver and bronze columns
del gold_usa['silver']
del gold_usa['bronze']
del gold_usa['id']
del gold_usa['dob']
del gold_usa['height']
del gold_usa['weight']
```

```python
#Add the year column to the DataFrame
year=2016
gold_usa["year"] = year
gold_usa
```

```python
#Reset the index
gold_usa.reset_index(inplace=True, drop=True)
gold_usa
```

```python
#Change the column name from gold to medal
gold_usa = gold_usa.rename(columns={"gold":"medal"})
gold_usa
```

```python
#Change the datatype to object 
gold_usa = gold_usa.astype({'medal': 'object'})
```

```python
gold_usa.dtypes
```

```python
#Replace data from 1 to gold 
gold_usa['medal'] = gold_usa['medal'].replace([1],'Gold')
```

```python
gold_usa
```

```python
summer_gold_usa
```

```python
#Get rid of unused columns
del summer_gold_usa['City']
del summer_gold_usa['Discipline']
del summer_gold_usa['Country']
del summer_gold_usa['Event']
del summer_gold_usa['Event_gender']
summer_gold_usa
```

```python
#Change column names
summer_gold_usa = summer_gold_usa.rename(columns={"Athlete":"name"})
summer_gold_usa = summer_gold_usa.rename(columns={"Country_Code":"nationality"})
summer_gold_usa = summer_gold_usa.rename(columns={"Gender":"sex"})
summer_gold_usa = summer_gold_usa.rename(columns={"Year":"year"})
summer_gold_usa = summer_gold_usa.rename(columns={"Sport":"sport"})
summer_gold_usa = summer_gold_usa.rename(columns={"Medal":"medal"})
summer_gold_usa
```

```python
#Change variables values  
summer_gold_usa['sex'] = summer_gold_usa['sex'].replace(["Men"],'male')
summer_gold_usa['sex'] = summer_gold_usa['sex'].replace(["Women"],'female')
summer_gold_usa
```

```python
#Clean data a bit more
new = summer_gold_usa["name"].str.split(",", n=1, expand=True)
summer_gold_usa["Last_name"] = new[0]
summer_gold_usa["First_name"] = new[1]
summer_gold_usa.drop(columns = ["name"], inplace = True)
summer_gold_usa["Last_name"] = summer_gold_usa["Last_name"].str.capitalize()
summer_gold_usa
```

```python
#Grab first names and last names and combine them together again 
summer_gold_usa["name"] = summer_gold_usa["First_name"] +" " + summer_gold_usa["Last_name"]
del summer_gold_usa['Last_name']
del summer_gold_usa['First_name']

summer_gold_usa
```

```python
gold_usa
```

```python
#olympics = pd.concat([summer_gold_usa, gold_usa], axis = 0, join; str = "outer")
frames= [summer_gold_usa, gold_usa]
olympics = pd.concat(frames)
olympics.reset_index(inplace=True, drop=True)
olympics =olympics.astype({'year': 'int'})
olympics
```
